```python
import numpy as np
import pandas as pd

np.random.seed(42)

num_datos = 2500

data_usarios = {
    'id_usuario': range(1, num_datos + 1),
    'edad': np.random.randint(18, 75, size=num_datos),
    'genero': np.random.choice(['M', 'F'], size=num_datos),
    'ingreso_anual': np.random.randint(20000, 100000, size=num_datos)
}

df_usuarios = pd.DataFrame(data_usarios)
df_usuarios.to_csv('data_usuarios.csv', index=False)

data_vehiculos ={
    'id_vehiculo':range(1, num_datos + 1),
    'id_usuario':range(1, num_datos + 1),
    'anio_fabricacion':np.random.randint(2005, 2026, size=num_datos),
    'kilometraje':np.random.randint(10000, 250000, size=num_datos),
    'tipo_motor':np.random.choice(['GASOLINA', 'ELECTRICO','HIBRIDO'], size=num_datos),
    'cilindraje_motor':np.round(np.random.uniform(1.0,5.0, size=num_datos),1),
    'fecha_ultima_revision':pd.to_datetime('2020-01-01') + pd.to_timedelta(np.random.randint(0, 1500, size=num_datos), unit='D')
}

df_vehiculos = pd.DataFrame(data_vehiculos)
df_vehiculos.to_csv('data_vehiculos.csv', index=False)

data_ventas = {
    'id_venta': range(1, num_datos + 1),
    'id_vehiculo': range(1, num_datos + 1),
    'fecha_venta': pd.to_datetime('2020-01-01') + pd.to_timedelta(np.random.randint(0, 1500, size=num_datos), unit='D'),
    'precio_venta': np.round(np.random.uniform(10000, 500000, size=num_datos),2 )
}

df_ventas = pd.DataFrame(data_ventas)
df_ventas.to_csv('data_ventas.csv', index=False)

print("Los 3 archivos (usuarios, vehiculos, ventas) se crearon con exito")



```
